```python
from pycaret.datasets import get_data
dataSets = get_data('index')
```

```python
from pycaret.clustering import *
ClusteringParameters = setup(JewelleryDataSet, verbose=False)
```

## 1. K-Means Clustering

**1.1 No Data Preprocessing**

```python
#for c=3
from pycaret.clustering import *
print("************* For Cluster = 3 *************")
kMeanClusteringParameters3 = setup(JewelleryDataSet, verbose=False)
KMeanClusteringModel3 = create_model('kmeans', num_clusters=3)
kMeanPrediction3 = assign_model(KMeanClusteringModel3)
print(kMeanPrediction3)

#for c=4
print("************* For Cluster = 4 *************")
kMeanClusteringParameters4 = setup(JewelleryDataSet, verbose=False)
KMeanClusteringModel4 = create_model('kmeans', num_clusters=4)
kMeanPrediction4 = assign_model(KMeanClusteringModel4)
print(kMeanPrediction4)

#for c=5
print("************* For Cluster = 5 *************")
kMeanClusteringParameters5 = setup(JewelleryDataSet, verbose=False)
KMeanClusteringModel5 = create_model('kmeans', num_clusters=5)
kMeanPrediction5 = assign_model(KMeanClusteringModel5)
print(kMeanPrediction5)
```

```python
plot_model(KMeanClusteringModel3, plot = 'elbow')
```

```python
plot_model(KMeanClusteringModel3, plot = 'silhouette')
```

```python
plot_model(KMeanClusteringModel3, plot = 'tsne')
```

**1.2 Using Normalization**

```python
setup(data = JewelleryDataSet, normalize = True, normalize_method = 'zscore', verbose=False)

print("For Cluster = 3")
norm3= create_model('kmeans', num_clusters = 3)

print("For Cluster = 4")
norm4= create_model('kmeans', num_clusters = 4)

print("For Cluster = 5")
norm5= create_model('kmeans', num_clusters = 5)
```

```python
plot_model(norm3, plot = 'elbow')
```

```python
plot_model(norm5, plot = 'silhouette')
```

```python
plot_model(norm5, plot = 'tsne')
```

**1.3 Using Tranformation**

```python
setup(data = JewelleryDataSet, transformation = True, transformation_method = 'yeo-johnson', verbose=False)

print("For Cluster = 3")
trans3 = create_model('kmeans', num_clusters = 3)

print("For Cluster = 4")
trans4 = create_model('kmeans', num_clusters = 4)

print("For Cluster = 5")
trans5 = create_model('kmeans', num_clusters = 5)
```
